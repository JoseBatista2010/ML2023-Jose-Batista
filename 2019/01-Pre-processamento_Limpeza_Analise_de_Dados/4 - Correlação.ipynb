# 4 - Correlação
**Direto ao ponto**: correlação é o coeficiente de interdependência entre duas ou mais variáveis.

Em muitos casos, quando analisamos algumas variáveis, essa interdependência é muito clara. Exemplo:
- Massa corporal e altura,
- Consumo de combustível e distância percorrida,
- Número de anúncios e produtos vendidos,
- Tempo de estudo e desempenho na avaliação.

Entretanto, algumas vezes essa correlação pode não ser muito clara.

**Correlação Positiva**: ocorre quando há uma concentração dos pontos em tendência crescente, ou seja, conforme a variável independente aumenta, a dependente também aumenta. Seria o caso do exemplo da venda de guarda-chuvas, mencionado acima.

**Correlação Negativa**: ocorre quando há uma concentração dos pontos em tendência decrescente, ou seja, conforme a variável independente aumenta, a variável dependente diminui. Por exemplo: quanto mais pessoas vacinadas contra o vírus da gripe, menor a taxa de pessoas infectadas.

**Correlação Nula**: ocorre quando os pontos não seguem uma tendência positiva nem negativa, há uma dispersão entre os pontos. Isso significa que não há correlação aparente entre as variáveis.

## Correlações positivas

Gera uma seed para ramdomizar. Cria um vetor x e o popula com mil valores aleatórios entre 0 e 50. Gera uma curva Gaussiana entre x e um y com 1000 valores entre 0 e 100.

```python
import numpy as np

np.random.seed(1)

# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Positive Correlation with some noise
y = x + np.random.normal(0, 100, 1000)

np.corrcoef(x, y)
```

Plota um gráfico mostrando a corelação entre x e y.

```python
# import matplotlib
import matplotlib.pyplot as plt

plt.scatter(x, y)
```

## Correlações negativas

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Negative Correlation with some noise
y = 100 - x + np.random.normal(0, 5, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

## Correlação fraca ou sem correlação

```python
x = np.random.randint(0, 500, 1000)
y = np.random.randint(0, 50, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```
